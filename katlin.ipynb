```python
Q1=df['fire_size'].quantile(0.25)
Q1
```

```python
Q3=df['fire_size'].quantile(0.75)
Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
lower_lim=Q1-(1.5*IQR)
upper_lim=Q3+(1.5*IQR)
```

```python
lower_lim
```

```python
upper_lim
```

```python
outliers_15_low=(df["fire_size"]<lower_lim)
outliers_15_up=(df["fire_size"]>upper_lim)
```

```python
len(df['fire_size'])-(len(df['fire_size'][outliers_15_low])+len(df["fire_size"][outliers_15_up]))
```
