```python
df['medium_plus']=''
```

```python
df.loc[df['fire_size_bin_no']>=4,'medium_plus']=1
df.loc[df['fire_size_bin_no']<4,'medium_plus']=0
df['medium_plus']=df["medium_plus"].astype(int)
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['discovery_month_no'] = le.fit_transform(df["discovery_month"])
df.head()
```

```python
# Months dictionary
months_num = {
    "Jan": 1,
    "Feb": 2,
    "Mar": 3,
    "Apr": 4,
    "May": 5,
    "Jun": 6,
    "Jul": 7,
    "Aug": 8,
    "Sep": 9,
    "Oct": 10,
    "Nov": 11,
    "Dec": 12,
}

# Months' names encoded using the dictionary values
df["discovery_month_no"] = df["discovery_month"].apply(lambda x: months_num[x])
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['state_no'] = le.fit_transform(df["state"])
df
```
