# First Round of Cleaning

## Import Initial Dependency & Read Original CSV

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Resources_k/Original Data/FW_Veg_Rem_Combined_ORIGINAL.csv",index_col=False)
df
```

```python
df.columns
```

### Rename Unnamed column to fire_id to create unique Primary Key value for database/

```python
df.drop(columns=["Unnamed: 0.1"], inplace=True)
```

```python
df.rename(columns={"Unnamed: 0": "fire_id"}, inplace=True)

```

```python
df
```

```python
df.describe()
```

```python
df.fire_size_class.value_counts()
```

### Drop Fire Name due to large quantity of fires w/out names

```python
df=df.drop(columns="fire_name")
df
```

### Drop Rows with Nulls

```python
df=df.dropna()
df
```

```python
df_new=df[['fire_id', 'fire_size',
       'fire_size_class', 'stat_cause_descr', 'latitude', 'longitude', 'state',
       'disc_clean_date', 'cont_clean_date', 'discovery_month',
       'putout_time',  'Vegetation','fire_mag', 'Temp_pre_30', 'Temp_pre_15', 'Temp_pre_7',
       'Temp_cont', 'Wind_pre_30', 'Wind_pre_15', 'Wind_pre_7', 'Wind_cont',
       'Hum_pre_30', 'Hum_pre_15', 'Hum_pre_7', 'Hum_cont', 'Prec_pre_30',
       'Prec_pre_15', 'Prec_pre_7', 'Prec_cont', 'remoteness']]
df_new
```

### Get Year from Fire Discovery Date

```python
df_new['disc_clean_date'] = pd.to_datetime(df_new['disc_clean_date'].str.strip(), format='%m/%d/%Y')
df_new['year'] = df_new.disc_clean_date.dt.year
```

```python
df_new.info()
```

```python
df_new
```

### Clean Irrelevant Columns (fire_mag and fire_class_size overlap with fire_size column)

```python
df_new.drop(["disc_clean_date", "cont_clean_date", "fire_mag", "fire_size_class"], axis=1, inplace=True)
```

```python
df_new
```

### Change Putout Time to integer 

```python
df_new["putout_time"].value_counts
```

```python
df_new["putout_time"]=df_new["putout_time"].map(lambda x : 0 if x[:6]=="0 days" else x)
df_new
```

```python
df_new.loc[df_new["putout_time"]==0]
```

```python
df_new["putout_time"].unique()
```

```python
# Adding these columns to the existing dataframe.
df_new[['A', 'B']] = df_new['putout_time'].str.split(' ', 1, expand=True)

df_new
```

```python
df_new["A"].unique()
```

```python
df_new[df_new["A"].isnull()]
```

```python
df_new['A'] = df_new['A'].replace(np.nan, 0)
df_new
```

```python
df_new.dtypes
```

```python
df_new = df_new.astype({"A":'int64'}) 
df_new.dtypes
```

```python
df_new.drop(["putout_time", "B"], axis=1, inplace=True)
df_new
```

```python
df_new.rename(columns = {'A':'putout_time'}, inplace=True)
df_new
```

### Change name of fire cause column to more relevant name

```python
df_new.rename(columns = {'stat_cause_descr':'fire_cause'}, inplace = True)
df_new
```

### Drop Vegetation and Remoteness columns due to lack of relevance

```python
df_new=df_new.drop(columns="Vegetation")
```

```python
df_new.columns
```

```python
df_new.drop(columns="remoteness", axis=1, inplace=True)
df_new
```

```python
df_new.info()
```

### Drop rows with temps of -1.000 due to preceived concern about accuracy

```python
df_new.loc[(df["Temp_pre_30"]!=-1.0)]
```

```python
df_new= df_new[df_new.Temp_pre_30 != -1.0]
df_new
```

```python
df.loc[(df["fire_cause"] == "Campfire") |
       (df["fire_cause"] == "Children") | 
       (df["fire_cause"] == "Smoking") | 
       (df["fire_cause"] == "Fireworks"), "fire_cause"] = "Accidental"

df
```

```python
df.fire_cause.value_counts()
```

### Drop fires with Fire Cause of "Other" (original categories of Miscellaneous, Missing/Undefined, Structure)

# Fourth Round of Cleaning: Binning Fire Size

### Binning Fire Size into more reasonable categories

```python
df
```

```python
df.describe()
```

```python
df["fire_size_bin"]=''
df
```

```python
df.loc[(df["fire_size"].between(0,5)), "fire_size_bin"] = "Teacup"
df.loc[(df["fire_size"].between(5,10)), "fire_size_bin"] = "Toy"
df.loc[(df["fire_size"].between(10,50)), "fire_size_bin"] = "Mini"
df.loc[(df["fire_size"].between(50,1000)), "fire_size_bin"] = "Medium"
df.loc[(df["fire_size"].between(1000,10000)), "fire_size_bin"] = "Large"
df.loc[(df["fire_size"]>10000), "fire_size_bin"] = "XL"
```

```python
#df.to_csv("Resources_k/fire_size_bins.csv", index=False)
```

```python
df['fire_size_bin_no']=''
```

```python
df.loc[(df["fire_size_bin"]=="Teacup"), "fire_size_bin_no"] = 1
df.loc[(df["fire_size_bin"]=="Toy"), "fire_size_bin_no"] = 2
df.loc[(df["fire_size_bin"]=="Mini"), "fire_size_bin_no"] = 3
df.loc[(df["fire_size_bin"]=="Medium"), "fire_size_bin_no"] = 4
df.loc[(df["fire_size_bin"]=="Large"), "fire_size_bin_no"] = 5
df.loc[(df["fire_size_bin"]=="XL"), "fire_size_bin_no"] = 6
```
